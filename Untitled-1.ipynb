```python
def gibbs_sampler(mu_s1_hyper = 4, mu_s2_hyper = -2, var_s1_hyper = 5, var_s2_hyper = 5, iterations = 1000):
    # Initilize
    A = np.array([[1, -1]]) # 2x1 column vector

    #Hyper
    sig_s1 = var_s1_hyper
    sig_s2 = var_s2_hyper
    sig_s = np.array([[sig_s1, 0], [0, sig_s2]])# 2x2 matrix
    sig_ts = 5
    #------ 
    t = 3
    mu_s = np.array([[mu_s1_hyper], [mu_s2_hyper]])   # 1x2 row vector
    variance_t = 5 # HYPER
    mu_t = 6 # Hyper mu_s1 - mu_s2

    s1_vec = []
    s2_vec = []
    t_vec = []

    for i in range(iterations):
        print(i)
        print(sig_s)
    # Corrected matrix multiplication for sig_st computation
     #   sig_st = np.linalg.inv(np.linalg.inv(sig_s) + (A * (1/sig_ts)) * A.T)
        sig_st = np.linalg.inv(np.linalg.inv(sig_s) + np.matmul(A.T*1/sig_ts, A))
        print(sig_st)
    # Compute mu_st
     #   mu_st = sig_st * (np.linalg.inv(sig_s) * mu_s + (A * (1/sig_ts)) * t)  
        mu_st = np.matmul(sig_st, np.matmul(np.linalg.inv(sig_s),mu_s) + A.T/sig_ts*t)
        mu_st_in = mu_st.flatten()
        print(f"mu_st {mu_st}")
        print(f"sig_st {sig_st}")
        s1,s2 = np.random.multivariate_normal(mu_st_in, sig_st) #Gives s1 and s2

        s1_vec.append(s1)
        s2_vec.append(s2)

        if i != 0:
            sig_s1 = np.var(s1_vec)
            sig_s2 = np.var(s2_vec)
            sig_s = np.array([[sig_s1, 0], [0, sig_s2]])


        mu_s=np.array([[np.mean(s1_vec)], [np.mean(s2_vec)]]) 
        mu_t = np.mean(s1_vec) - np.mean(s2_vec) #Defined as mu_s1 - mu_s2
        if mu_t > 0:
            t = np.random.normal(mu_t,variance_t)
        else:
            t = 0

        t_vec.append(t)
        if i != 0:
            sig_ts = np.var(t_vec)      #How does these differ?
            variance_t = np.var(t_vec)  #How does these differ? wtf

    return s1_vec, s2_vec, t_vec
```

```python
#  Den funkar ibland!

def post_erior_gibbs_sampler_estimate(s1,s2,t):
    mu_s=np.array([[np.mean(s1)], [np.mean(s2)]])
    covariance_s = np.cov(s1,s2)
    variance_s1 = np.var(s1)
    variance_s2 = np.var(s2)

    return mu_s, covariance_s, variance_s1,variance_s2

```

```python
# Define the mean vector and the covariance matrix

mean, covariance_matrix,variance_s1,variance_s2 = post_erior_gibbs_sampler_estimate(s1,s2,t)
mean = mean.flatten()

print(f"Mean for s1 and s2: {mean}")
print(f"Covariance for s1 and s2: {covariance_matrix}")

# Create a grid of points at which to evaluate the Gaussian
x = np.linspace(2, 6, 100)
y = np.linspace(-4, 0, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Evaluate the Gaussian at each grid point
rv = multivariate_normal(mean, covariance_matrix)
Z = rv.pdf(pos)

# Create a contour plot of the Gaussian
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multivariate Gaussian Distribution Contour Plot')
plt.show()


# Set up plot
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(270, 45)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(Z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)
```

```python
def prior_vs_posterior_comp(mu_1_hyper, mu_2_hyper, var_s1_hyper, var_s2_hyper,mu_s,variance_s1,variance_s2):
    mu_s1 = mu_s[0]
    mu_s2 = mu_s[1]

    # Axis [mean-3*std, mean + 3*std]
    x_axis_s1 = np.linspace(mu_s1 - 3*np.sqrt(variance_s1),      mu_s1 + 3*np.sqrt(variance_s1),     100 )
    x_axis_s2 = np.linspace(mu_s2 - 3*np.sqrt(variance_s2),      mu_s2 + 3*np.sqrt(variance_s2),     100 )

    pdf_s1_values_gs =  norm.pdf(x_axis_s1,     mu_s1   ,   np.sqrt(variance_s1))
    pdf_s1_values_hyp = norm.pdf(x_axis_s1,     mu_1_hyper, np.sqrt(var_s1_hyper))

    pdf_s2_values_gs =  norm.pdf(x_axis_s2,     mu_s2   ,   np.sqrt(variance_s2))
    pdf_s2_values_hyp = norm.pdf(x_axis_s2,     mu_2_hyper, np.sqrt(var_s2_hyper))

    #Prior and after s1
    plt.plot(x_axis_s1, pdf_s1_values_gs,  label=f'Distrb for S1 after GS')
    plt.plot(x_axis_s1, pdf_s1_values_hyp, label=f'Distrb for S2 before GS')
    plt.legend(loc='upper right')
    plt.grid()
    plt.xlabel('s1')
    plt.ylabel('Probability Density')
    plt.show()

    #Prior and after s2
    plt.plot(x_axis_s2, pdf_s2_values_gs,label=f'Distrb for S2 after GS')
    plt.plot(x_axis_s2, pdf_s2_values_hyp, label=f'Distrb for S2 before GS')
    plt.legend(loc='upper right')
    plt.grid()
    plt.xlabel('s2')
    plt.ylabel('Probability Density')
    plt.show()

# One could implemented so that we have a hyperparameter function!
mu_s1_hyper = 4
mu_s2_hyper = -2 
var_s1_hyper = 5 
var_s2_hyper = 5 


mu_s,_,variance_s1, variance_s2 = post_erior_gibbs_sampler_estimate(s1,s2,t)
prior_vs_posterior_comp(mu_s1_hyper,mu_s2_hyper,var_s1_hyper,var_s2_hyper,mu_s, variance_s1, variance_s2)
```
